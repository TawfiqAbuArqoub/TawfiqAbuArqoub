## **Load and Inspect Data**



```python
import pandas as pd
import numpy as np

url = '/content/sales_predictions_2023 - sales_predictions_2023.csv'
df = pd.read_csv(url)
df.info()
df.head()
```

## **Clean Data**

```python
df['Item_Weight'].fillna(df['Item_Weight'].mean(), inplace=True)
df['Outlet_Size'].fillna('Medium', inplace=True)
df.info()
```

## **Expolatory Data Analysis**

```python
df = pd.DataFrame(df)
# Number of rows and columns
num_rows ,num_cols = df.shape
print(f'Number of rows: {num_rows}, Number of columns: {num_cols}')
```

```python
print(df.dtypes) #datatype of each variable
```

```python
# check for duplicate rows
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
```

```python
#Identify missing values.
missing_values = df.isna().sum()
missing_values
```

```python
# address missing values with placeholder value
df['Item_Weight'].fillna(-1, inplace=True)
df['Outlet_Size'].fillna('M', inplace=True)
```

```python
# confirm that there are no missing values after addressing them.
df.isna().sum()
```

```python
#Find and fix any inconsistent categories of data
df['Item_Fat_Content'] = df['Item_Fat_Content'].replace({'LF': 'Low Fat', 'low fat': 'Low Fat', 'reg': 'Regular'})
df['Item_Fat_Content'].value_counts()
```

```python
# Summary statistics for numerical columns
sum_stats = df.describe()
sum_stats
```
